```python
# Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Importing dataset

df = pd.read_csv('archive/Amazon Customer Behavior Survey.csv')
```

Let's showcase a distribution plot to further understand the behaviour of these variables. 

```python
# Plotting bar plots for categorical dimensions

categorical_dimensions = ['Purchase_Frequency', 
                          'Personalized_Recommendation_Frequency', 
                          'Browsing_Frequency', 
                          'Saveforlater_Frequency', 
                          'Review_Reliability', 
                          'Recommendation_Helpfulness', 
                          'Review_Helpfulness',
                          'Search_Result_Exploration', 
                          'Service_Appreciation'
                          ]

for dimension in categorical_dimensions:
    # Figure size
    fig, ax = plt.subplots(figsize =(16, 6))
    
    # Declaring axis    
    keys = df[dimension].unique()
    values = df[dimension].value_counts()
    
    # Remove x, y Ticks
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')
    
    # Remove axes splines
    for s in ['top', 'bottom', 'left', 'right']:
        ax.spines[s].set_visible(False)
        
    # Setting title
    ax.set_title('Bar plot for dimension ' + dimension, loc ='left', )
    
    # Show top values
    ax.invert_yaxis()
    
    # Showcasing the result
    plt.barh(keys, values)
    plt.show()
```
